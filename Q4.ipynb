# Imagine you working as a sale manager now you need to predict the Revenue and whether that particular revenue is on the weekend or not and find the Informational_Duration using the Ensemble learning algorithm

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('online_shoppers_intention.csv')
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['VisitorType'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df['VisitorType'] = enc.fit_transform(df['VisitorType'])
```

```python
df.replace({False: 0, True: 1}, inplace=True)
```

```python
enc2 = LabelEncoder()
df['Month'] = enc2.fit_transform(df['Month'])
```

```python
sns.heatmap(df.corr())
```

```python
X = df.drop(['Weekend'], axis=1)
y = df['Weekend']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
#predicting Weekend
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#preedicting Revenue
X = df.drop(['Revenue'], axis=1)
y = df['Revenue']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
model2 = GradientBoostingClassifier()
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```
